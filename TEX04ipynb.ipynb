```python
lista_livro = []; #Declara uma lista para conter dicionários com livros
id_global = 0;  #Declara uma var global para implementar os ids dos livros

#Declara uma função para cadastrar livros
def cadastrar_livro(id):
  print('*' * 70);
  print('-' * 24, 'MENU CADASTRAR LIVRO', '-' * 24);
  print('id do livro {}'.format(id));
  nome = input('Por favor entre com o nome: ');
  autor = input('Por favor entre com o autor: ');
  editora = input('Por favor entre com a editora: ');
  lista_livro.append({'id':id, 'nome':nome, 'autor':autor, 'editora':editora});

#Declara uma função para consultar os livros previamente cadastrados
def consultar_livro():
  while True:
    print('*' * 70);
    print('-' * 24, 'MENU CONUSLTAR LIVRO', '-' * 24)

    print('Escolha a opção desejada:');
    print('1 - Consultar Todos os Livros');
    print('2 - Consultar Livro por id');
    print('3 - Consultar Livro(s) por autor');
    print('4 - Retornar');

    res = input('>> ');

    if (res != '1' and res != '2' and res != '3' and res != '4'):
      print('Opção inválida. Tente novamente.');
      continue;

    if (res == '1'):
      for livro in lista_livro:
        print('\nid: {}'.format(livro['id']));
        print('Nome: {}'.format(livro['nome']));
        print('Autor: {}'.format(livro['autor']));
        print('Editora: {}\n'.format(livro['editora']));
    elif (res == '2'):
      resId = int(input('\nDigite o id do livro: '));
      for livro in lista_livro:
        if (livro['id'] == resId):
          print('\nid: {}'.format(livro['id']));
          print('Nome: {}'.format(livro['nome']));
          print('Autor: {}'.format(livro['autor']));
          print('Editora: {}\n'.format(livro['editora']));
    elif(res == '3'):
      resAutor = input('\nDigite o autor do(s) livro(s): ');
      for livro in lista_livro:
        if (livro['autor'] == resAutor):
          print('\nid: {}'.format(livro['id']));
          print('Nome: {}'.format(livro['nome']));
          print('Autor: {}'.format(livro['autor']));
          print('Editora: {}\n'.format(livro['editora']));
    elif (res == '4'):
      break;

#Declara uma função para remover um determinado livro pelo seu id
def remover_livro():
  print('*' * 70);
  print('-' * 25, 'MENU REMOVER LIVRO', '-' * 25);
  resRemover = int(input('Digite o id do livro a ser removido: '));
  for livro in lista_livro:
    if (livro['id'] == resRemover):
      lista_livro.remove(livro);

#Mensagem de entrada
print('Bem-vindo(a) ao Controle de Livros do Matheus Vinicius Garcia Pombeiro\n');

#Programa principal (main)
while True:
  print('*' * 70);
  print('-' * 27, 'MENU PRINCIPAL', '-' * 27);
  print('Escolha a opção desejada:');
  print('1 - Cadastrar Livro');
  print('2 - Consultar Livro');
  print('3 - Remover Livro');
  print('4 - Sair');

  res = input('>> ');

  if (res != '1' and res != '2' and res != '3' and res != '4'):
    print('Opção inválida. Tente novamente.');
    continue;

  if (res == '1'):
    id_global += 1;
    cadastrar_livro(id_global);
  elif (res == '2'):
    consultar_livro();
  elif (res == '3'):
    remover_livro();
  elif(res == '4'):
    break;
```
