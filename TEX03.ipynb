```python
#Função que recebe e retorna o serviço
def escolha_servico():
  while True:
    print('Entre com o tipo de serviço desejado');
    print('DIG - Digitalização');
    print('ICO - Impressão Colorida');
    print('IPB - Impressão Preto e Branco');
    print('FOT - Fotocópia');
    servico = input('>> ');

    if (servico not in 'DIGdig' and servico not in 'ICOico' and servico not in 'IPBipb' and servico not in 'FOTfot'):
      print('\nEscolha inválida.');
      print('Entre com o tipo de serviço desejado novamente.\n');
      continue;

    if (servico in 'DIGdig'):
      valor = 1.1;
    elif (servico in 'ICOico'):
      valor = 1;
    elif (servico in 'IPBipb'):
      valor = 0.4;
    elif (servico in 'FOTfot'):
      valor = 0.2;

    return valor;
    break;

#Função que recebe e retorna a quantidade de páginas
def num_pagina():
  while True:
    try:
      numPaginas = int(input('\nEntre com o número de páginas: '));
    except ValueError:
      print('\nValor inválido.');
      print('Entre com a quantidade de páginas novamente.')
      continue;

    if(numPaginas >= 20000):
      print('\nNão aceitamos tantas páginas de uma vez.');
      print('Por favor, entre com o número de páginas novamente.');
      continue;

    if (numPaginas < 20):
      taxa = 0;
      desconto = numPaginas * taxa / 100;
    elif (numPaginas >= 20 and numPaginas < 200):
      taxa = 15;
      desconto = numPaginas * taxa / 100;
    elif (numPaginas >= 200 and numPaginas < 2000):
      taxa = 20;
      desconto = numPaginas * taxa / 100;
    elif (numPaginas >= 2000 and numPaginas < 20000):
      taxa = 25;
      desconto = numPaginas * taxa / 100;

    valor = numPaginas - desconto;
    return valor;
    break;

#Função que recebe e retorna o serviço extra
def servico_extra():
  while True:
    print('\nDeseja adicionar mais algum serviço?');
    print('1 - Encadernação Simples - R$ 15,00');
    print('2 - Encadernação Capa Dura - R$ 40,00');
    print('0 - Não desejo mais nada');
    servicoExtra = input('>> ');

    if (servicoExtra != '1' and servicoExtra != '2' and servicoExtra != '0'):
      print('\nValor inválido\n');
      print('Entre novamente com o serviço extra desejado.\n');
      continue;

    if (servicoExtra == '1'):
      valor = 15;
    elif (servicoExtra == '2'):
      valor = 40;
    elif (servicoExtra == '0'):
      valor = 0;

    return valor;
    break;

#Mensagem de entrada
print('Bem-vindo(a) à papelaria do Matheus Vinicius Garcia Pombeiro\n');

#Chama as funções e armazena os seus valores
servico = escolha_servico();
paginas = num_pagina();
servicoExtra = servico_extra();

#Processa os valores retornados das funções
valorTotal = servico * paginas + servicoExtra;

#Mensagem de saída final
print('\nTotal (R$): {:.2f} (serviço: {} * páginas: {} + extra(s): {})'.format(valorTotal, servico, paginas, servicoExtra));
```
